```python
map=Basemap(projection='mill',
           llcrnrlat=-40,       # lower left corner latitude
           llcrnrlon=-40,       #   ''   ''    ''   longitude
           urcrnrlat=50,        # upper right  ''   latitude
           urcrnrlon=75,        #   ''   ''    ''   longitude
           resolution='c')      # c=crude, l=low, h=high, f=full
map.drawcoastlines()
map.drawcountries(linewidth=1)
map.fillcontinents()
plt.show()
```

```python
map=Basemap(projection='mill',
           llcrnrlat=-90,       # lower left corner latitude
           llcrnrlon=-180,       #   ''   ''    ''   longitude
           urcrnrlat=90,        # upper right  ''   latitude
           urcrnrlon=180,        #   ''   ''    ''   longitude
           resolution='c')      # c=crude, l=low, h=high, f=full
map.drawcoastlines()
map.drawcountries(linewidth=1)
map.drawstates(color='b')
map.drawrivers()
#map.drawmapboundary()    # not required
# map.drawcounties()
#map.fillcontinents()
#map.etopo()
#map.bluemarble()
plt.show()
```

```python
# plotting coordinates and connecting lines
```

```python
m=Basemap(projection='mill',
           llcrnrlat=-25,       
           llcrnrlon=-130,       
           urcrnrlat=50,        
           urcrnrlon=-60,       
           resolution='c')
m.drawcoastlines()
m.drawcountries(linewidth=1)
m.drawstates(color='b')

# connecting both coordinates
xs=[]
ys=[]

# 1st coordinate NewYork
NYClat,NYClon=40.7127,-74.0059
xpt,ypt=m(NYClon,NYClat)
xs.append(xpt)
ys.append(ypt)
m.plot(xpt,ypt,'r*',markersize=15)

# 2nd coordinate LosAngelos
LAlat,LAlon=34.05,-118.25
xpt,ypt=m(LAlon,LAlat)
xs.append(xpt)
ys.append(ypt)
m.plot(xpt,ypt,'b^',markersize=15)

m.plot(xs,ys,color='red',linewidth=1,label='Connect')

m.drawgreatcircle(NYClon,NYClat,LAlon,LAlat,color='c',linewidth=2, label='Arc')   # for drawing curves between two points

plt.legend(loc=4)
plt.show()

```

```python
from numpy import array
lat=data['earthquake.latitude'].head(20)
lon=data['earthquake.longitude'].head(20)
lat=array(lat)
lon=array(lon)
```

```python
world_map=Basemap(projection='mill',
                  llcrnrlat=-90,       
                  llcrnrlon=-180,       
                  urcrnrlat=90,        
                  urcrnrlon=180,        
                  resolution='c')      
world_map.drawcoastlines()
world_map.drawcountries(linewidth=1)
world_map.drawstates(color='b')

x,y=world_map(lon,lat)
world_map.plot(x,y,'ro',markersize=15,alpha=0.5)

plt.show()

```
