```python
from numpy import *
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
data=pd.read_csv(r'C:\Users\Brij\Desktop\Python for data science\2019.csv')
data1=data.copy()
```

```python
data1.columns
```

```python
data1['Overall rank']=data1['Overall rank'].astype('float64')
```

```python
X=data1.drop(['Overall rank','Country or region'],axis=1).values
y=data1['Overall rank'].values
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
reg=LinearRegression()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
reg.fit(X_train,y_train)
```

```python
prediction=reg.predict(X_test)
```

```python
coeff_df = pd.DataFrame(reg.coef_)
```

```python
coeff_df
```

```python
from sklearn.metrics import accuracy_score
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': prediction})
df
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, prediction))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, prediction))  
print('Root Mean Squared Error:',sqrt(metrics.mean_squared_error(y_test, prediction)))
```
